```python
def tmaker(df):
    teams=[]
    for i, row in df.iterrows():
        if df['HomeTeam'][i] not in teams:
            teams.append(df['HomeTeam'][i])
    return (teams)
    
def fmaker(df,teams):
    forms=[]
    for team in teams:
        n=0
        st=""
        end = len(df) -1
        for i in range(end,0,-1):
            if df['HomeTeam'][i] == team and df['FTR'][i] == 'H':
                st='W '+st
                n+=1
            elif df['HomeTeam'][i] == team and df['FTR'][i] == 'D':
                st='D '+st
                n+=1
            elif df['HomeTeam'][i] == team and df['FTR'][i] == 'A':
                st='L '+st
                n+=1
            elif df['AwayTeam'][i] == team and df['FTR'][i] == 'A':
                st='W '+st
                n+=1
            elif df['AwayTeam'][i] == team and df['FTR'][i] == 'D':
                st='D '+st
                n+=1
            elif df['AwayTeam'][i] == team and df['FTR'][i] == 'H':
                st='L '+st
                n+=1
            if n==7:
                break
        forms.append(st)
    return (forms)
```

```python
L=["https://www.football-data.co.uk/mmz4281/1920/E0.csv", "https://www.football-data.co.uk/mmz4281/1920/P1.csv", "https://www.football-data.co.uk/mmz4281/1920/B1.csv", "https://www.football-data.co.uk/mmz4281/1920/N1.csv", "https://www.football-data.co.uk/mmz4281/1920/F2.csv", "https://www.football-data.co.uk/mmz4281/1920/F1.csv", "https://www.football-data.co.uk/mmz4281/1920/SP2.csv", "https://www.football-data.co.uk/mmz4281/1920/SP1.csv", "https://www.football-data.co.uk/mmz4281/1920/I2.csv", "https://www.football-data.co.uk/mmz4281/1920/I1.csv", "https://www.football-data.co.uk/mmz4281/1920/D2.csv", "https://www.football-data.co.uk/mmz4281/1920/E1.csv", "https://www.football-data.co.uk/mmz4281/1920/D1.csv", "https://www.football-data.co.uk/mmz4281/1920/D2.csv"]
teams = forms = []
for url in  L:
    #print (url)
    s = requests.get(url).content
    df = pd.read_csv(io.StringIO(s.decode('utf-8')))
    
    T = tmaker(df)
    F = fmaker(df,T)
    teams=teams+T
    forms=forms+F
```
