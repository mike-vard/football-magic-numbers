```python
#PROBABILTY MAKER FOR FUTURE GAMES

import numpy as np  
import pandas as pd 
import io
import requests
```

```python
url = "http://api.clubelo.com/Fixtures"
s = requests.get(url).content
df = pd.read_csv(io.StringIO(s.decode('utf-8')))

df = df[['Date','Country', 'Home','Away', 'GD<-5','GD=-5','GD=-4','GD=-3','GD=-2','GD=-1','GD=0','GD=1','GD=2','GD=3','GD=4','GD=5','GD>5']]
df['A']= df.iloc[:, -13:-7].sum(axis=1)  
df['H']= df.iloc[:, -7:-1].sum(axis=1)    

df.drop(['GD<-5','GD=-5','GD=-4','GD=-3','GD=-2','GD=-1','GD=1','GD=2','GD=3','GD=4','GD=5','GD>5'], axis=1, inplace=True)
df.rename(columns={'GD=0':'D'}, inplace=True)
df = df[['Date','Country', 'Home','Away', 'H', 'D', 'A']]
```
