```python
for index, row in df.iterrows():
    L=[]
    n=0
    for char in df['FormH'][index]:
        if char in ["W","D","L"]:
            n+=1
            L.append(char)
    df.at[index, 'FormH'] = L
for index, row in df.iterrows():
    L=[]
    n=0
    for char in df['FormA'][index]:
        if char in ["W","D","L"]:
            n+=1
            L.append(char)
    df.at[index, 'FormA'] = L

for index, row in df.iterrows():
    for char in df['Result'][index]:
        if char in ["H","D","A"]:
            df.at[index, 'Result'] = char
```

```python
for index, row in df.iterrows():
    df.at[index, 'DivSum'] = df['Sum'][index]/float(df['SumA'][index]+0.5)
    df.at[index, 'DSum'] = df['Sum'][index]-(df['SumA'][index])
```

```python
#prob maker1

for i, row in df.iterrows():
    #h = df['Sum'][i]+3
    #a = df['SumA'][i]+3
    #d = int((a+h)*(26)/72)
    if df['Result'][i]=='H':
        h = int(df['EloH'][i]/100)+20
        a = int(df['EloA'][i]/100)+10
        d = int((a+h)*(26)/72)
        total = a+h+d
        df.at[i, 'MP_H'] = h/float(total)
        df.at[i, 'MP_D'] = d/float(total)
        df.at[i, 'MP_A'] = a/float(total)
    elif df['Result'][i]=='A':
        a = int(df['EloH'][i]/100)+20
        h = int(df['EloA'][i]/100)+10
        d = int((a+h)*(26)/72)
        total = a+h+d
        df.at[i, 'MP_H'] = h/float(total)
        df.at[i, 'MP_D'] = d/float(total)
        df.at[i, 'MP_A'] = a/float(total)
    else:
        a = int(df['EloH'][i]/100)
        h = int(df['EloA'][i]/100)
        d = int((a+h)*(26)/72)+3
        total = a+h+d
        df.at[i, 'MP_H'] = h/float(total)
        df.at[i, 'MP_D'] = d/float(total)
        df.at[i, 'MP_A'] = a/float(total)
        

```
